```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Firstly we have to extract the data sets and join them accodingly:
customer=pd.read_csv("D:/Projects/Python Foundation Case Study 2 - Credit Card Case Study B/Customer Acqusition.csv")
tab_repay=pd.read_csv("D:/Projects/Python Foundation Case Study 2 - Credit Card Case Study B/Repayment.csv")
tab_spend=pd.read_csv("D:/Projects/Python Foundation Case Study 2 - Credit Card Case Study B/Spend.csv")
```

```python
# We drops irrelevant variable:
tab_repay.drop(columns='Unnamed: 4',inplace=True)
```

```python
# We drops missing values observations:
tab_repay.dropna(subset='Month',inplace=True)
tab_repay.Month
```

```python
# We have to convert date variables from object into datetime datatypes:
tab_repay.Month=pd.to_datetime(tab_repay.Month,format='%d-%b-%y')
tab_spend.Month=pd.to_datetime(tab_spend.Month,format='%d-%b-%y')
```

## 1. In the above dataset,

###  a. In case age is less than 18, replace it with mean of age values.

```python
customer.Age=pd.Series(np.where(customer.Age<18,customer.Age.mean(),customer.Age))
```

```python
# Now we have to merge data sets as per requirement:
spend=pd.merge(left=customer,right=tab_spend,how='inner',left_on='Customer',right_on='Customer',)
repay=pd.merge(left=customer,right=tab_repay,how='inner',left_on='Customer',right_on='Customer',)
```

###  b. In case spend amount is more than the limit, replace it with 50% of that customer’s limit. (customer’s limit provided in acquisition table is the per transaction limit on his card) 

```python
spend.loc[spend.Amount>spend.Limit,'Amount']=spend.Limit/2
```

###  c. Incase the repayment amount is more than the limit, replace the repayment with the limit. 

```python
repay.loc[repay.Amount>repay.Limit,"Amount"]=repay.Limit
```

## 2. From the above dataset create the following summaries:

### a. How many distinct customers exist? 

```python
pd.Series(spend.Customer.unique()).count()
```

### b. How many distinct categories exist? 

```python
pd.Series(spend.Type.unique()).count()
```

###  c. What is the average monthly spend by customers? 

```python
spend.groupby([spend.Month.dt.month_name(),'Customer']).Amount.mean()
```

### d. What is the average monthly repayment by customers? 

```python
repay.groupby([repay.Month.dt.month_name(),'Customer']).Amount.mean()
```

###  e. If the monthly rate of interest is 2.9%, what is the profit for the bank for each month? (Profit is defined as interest earned on Monthly Profit. Monthly Profit = Monthly repayment – Monthly spend. Interest is earned only on positive profits and not on negative amounts)

```python
# Firstly, we have to create profit variable by given formula:
repay['monthly_profit']=repay.Amount-spend.Amount
```

```python
# Now,create 2.9% earned profit variable:
repay['earned_profit']=round(repay.loc[repay.monthly_profit>0,'monthly_profit']*0.029,2)
```

```python
repay.groupby([repay.Month.dt.year,repay.Month.dt.month_name()]).earned_profit.sum()
```

###  f. What are the top 5 product types? 

```python
spend.groupby('Type').Amount.sum().sort_values(ascending=False).head(5)
```

### g. Which city is having maximum spend? 

```python
spend.groupby('City').Amount.sum().sort_values(ascending=False).head(1)
```

###  h. Which age group is spending more money? 

```python
# Firstly,we have to create age group:
spend['age_group']=pd.Series(np.where((spend.Age>17)&(spend.Age<36),'Young',np.where((spend.Age>35)&(spend.Age<60),
                                                                                     'Adult','Senior')))
```

```python
spend.groupby('age_group').Amount.sum().sort_values(ascending=False).head(1)
```

## 3. Calculate the city wise spend on each product on yearly basis. Also include a graphical representation for the same.


```python
city_wise=pd.crosstab(index=spend.City,columns=[spend.Product,spend.Month.dt.year],values=spend.Amount,aggfunc='sum')
city_wise.plot(kind='bar',figsize=[13,6])
plt.ylabel('Amount in Millions')
plt.show()
```

## 4. Create graphs for

### a. Monthly comparison of total spends, city wise

```python
city_comp=pd.crosstab(index=spend.Month.dt.month_name(),columns=spend.City,values=spend.Amount,aggfunc='sum')
city_comp.plot(kind='bar',figsize=[13,6])
plt.ylabel('total_amt (in Millions)')
plt.show()
```

### b. Comparison of yearly spend on air tickets

```python
spend.loc[spend.Type=='AIR TICKET',].groupby(spend.Month.dt.year).Amount.sum().plot(kind='bar',figsize=[10,6])
plt.xlabel('Yearly spend on AIR TICKET')
plt.ylabel('spending amount in millions')
plt.show()
```

### c. Comparison of monthly spend for each product (look for any seasonality that exists in terms of spend)

```python
pd.crosstab(index=[spend.Month.dt.month],columns=spend.Product,values=spend.Amount,aggfunc='sum').plot(kind='line',figsize=[13,6])
plt.show()
# After visualize the data, we can see the "higher spending" in months(January,May and November).
```

## 5. Write user defined PYTHON function to perform the following analysis: 

### You need to find top 10 customers for each city in terms of their repayment amount by different products and by different time periods i.e. year or month. The user should be able to specify the product (Gold/Silver/Platinum) and time period (yearly or monthly) and the function should automatically take these inputs while identifying the top 10 customers.


```python
def top10(*value):
    product=input("Enter the type of product").capitalize()
    period=input("Enter the period type year/month").lower()
    if period=='year':
        func=repay.loc[repay.Product==product].groupby([repay.Month.dt.year,'City','Customer']).Amount.sum()
    elif period=='month':
        func=repay.loc[repay.Product==product].groupby([repay.Month.dt.month,'City','Customer']).Amount.sum()
    else:
        print("Enter correct words")
    return func.reset_index().head(10)
```
